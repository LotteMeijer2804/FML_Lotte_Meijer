```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

df1 = pd.read_csv('steps_data.csv', sep=";")
df1.head()
```

```python
df2 = pd.read_csv('survey_data.csv')
df2.head()
```

```python
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head()
```

```python
df.info() #Get information on the variables
```

```python
df['waist'].value_counts() #Get all the values. This may be less useful when there are many values
```

```python
# The handy Pandas function to_numeric converts non-numeric data to NaN. The last argument overrides the errors
# Python would normally generate.
# The apply method of a dataframe lets us apply a function to all the elements.
df['waist'] = df['waist'].apply(pd.to_numeric, errors='coerce')

#To remove the values below 5 we need to create a function that we then apply to the entire column
def below_25(x):
    if(x < 25.0): 
        return float('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(below_25)

#To remove the values above 140 we need to create a function that we then apply to the entire column
def above_140(x):
    if(x > 140.0): 
        return float('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(above_140)
```

```python
sns.distplot(df['waist'].dropna(), kde=False) #Selecting the waist column. I need to drop the NA's for the plot
plt.title('Waist')
plt.xlabel('Waist (cm)')
plt.show()
```

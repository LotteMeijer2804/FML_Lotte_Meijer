```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

df1 = pd.read_csv('steps_data.csv', sep=";") #the sep says were the values should be seperated
df1.head()
```

```python
df2 = pd.read_csv('survey_data.csv')
df2.head()
```

```python
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head()
```

```python
df.info() #Get information on the variables
```

```python
df['waist'].value_counts() #Get all the values. This may be less useful when there are many values
```

```python
# The handy Pandas function to_numeric converts non-numeric data to NaN. The last argument overrides the errors
# Python would normally generate.
# The apply method of a dataframe lets us apply a function to all the elements.
df['waist'] = df['waist'].apply(pd.to_numeric, errors='coerce')

#To remove the values below 5 we need to create a function that we then apply to the entire column
def below_25(x):
    if(x < 25.0): 
        return float('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(below_25)

#To remove the values above 140 we need to create a function that we then apply to the entire column
def above_140(x):
    if(x > 140.0): 
        return float('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(above_140)
```

```python
waist_number = df['waist'].dropna() #Selecting the waist column. I need to drop the NA's for the plot
sns.distplot(waist_number)
plt.title('Waist of people')
plt.xlabel('Waist (cm)')
plt.ylabel('Number of people')
plt.show()
```

About this distribution:

It has 1 peak (or mode - unimodal)
It doesn't really hava a tail, maybe a bit more on the left, so it's left-skewed
The median will be before the mean

```python
mean_waist = df["waist"].mean()
median_waist = df["waist"].median()

sns.distplot(waist_number, kde=False)
plt.title('The mean and median of the waist')
plt.axvline(median_waist, 0, 100, color='green', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.axvline(mean_waist, 0, 100, color='blue', label='mean') 
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```

```python
sns.boxplot(df['height'].dropna()) #Selecting the weight column. I need to drop the NA's for the plot
plt.title('Height of people')
plt.xlabel('Height (cm)')
plt.ylabel('Number of people')
plt.show()
```
